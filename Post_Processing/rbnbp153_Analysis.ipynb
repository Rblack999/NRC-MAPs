```python
import time
import numpy as np
import json
from impedance import preprocessing
from numpy import genfromtxt
from scipy import stats


def fit_EIS_csv(root_path,test_name,area,current_used,type):


    files = [#f'{root_path}{test_name}_Char_01_GEIS_C01.csv',
             f'{root_path}{test_name}_Char_03_GEIS_C01.csv',
             f'{root_path}{test_name}_Char_05_GEIS_C01.csv',
             f'{root_path}{test_name}_Char_07_GEIS_C01.csv',
             f'{root_path}{test_name}_Char_09_GEIS_C01.csv',
             f'{root_path}{test_name}_Char_11_GEIS_C01.csv',
             f'{root_path}{test_name}_Char_13_GEIS_C01.csv',
             f'{root_path}{test_name}_Char_15_GEIS_C01.csv',
             f'{root_path}{test_name}_Char_17_GEIS_C01.csv',
#         f'{root_path}{test_name}_Char_19_GEIS_C01.csv',
       #f'{root_path}{test_name}_Char_21_GEIS_C01.csv',
             ]
    #Take the values from the script Multi_EISFit.py
    #Categorize each GEIS based on step + timing - use global tracker?
    #Fit and extract the data
    #Store into a separate file + format, make a new metrics .json

    k = 0
    time_value = np.zeros(len(files))  #Can be voltages (V) or time (min) based on PEIS or GEIS
    #for i in range(len(files)):
    #     voltages[i] = k + 1.66 #100 seconds to minutes
    #     k = k + 1.66
    Rs = np.zeros(len(time_value))
    Rct = np.zeros(len(time_value))
    a_hat_1 = np.zeros(len(time_value))
    C1 = np.zeros(len(time_value))
    ECSA = np.zeros(len(time_value))
    j = 0 #increment counter only
    #etc.
    for i in files:
        data = genfromtxt(i, delimiter=',', skip_header=67)
        data = np.transpose(data)
        frequencies = data[0]
        if type == 'GEIS':
            abs_Ewe = data[11]
            abs_I = data[12]
            Phase_Zwe = data[4]
            t = np.mean(data[5])/60
            time_value[k] = t
            k = k+1
        if type == 'PEIS':
            abs_Ewe = data[12]
            abs_I = data[13]
            Phase_Zwe = data[4]

        #For completeness sake, put into a dictionary
        data = {'frequencies':frequencies.tolist(),'abs_Ewe':abs_Ewe.tolist(),'abs_I':abs_I.tolist(),'Phase_Zwe':Phase_Zwe.tolist()}

        abs_Z = np.divide(data['abs_Ewe'], data['abs_I'])
        Re_Z = abs_Z * (np.cos(np.radians(data['Phase_Zwe'])))
        Im_Z = abs_Z * (np.sin(np.radians(data['Phase_Zwe'])))
        #Convert Re_Z and Im_Z into a single complex numpy array
        Z = Re_Z + 1j*Im_Z

        #Below is built in library to cut frequencies below x-axis
        frequencies, Z = preprocessing.ignoreBelowX(frequencies,Z)
        mask = frequencies < 5000
        frequencies = frequencies[mask]
        Z = Z[mask]
        from impedance.models.circuits import CustomCircuit
        # A different circuit
        circuit = 'R0-p(R1,CPE1)'
        initial_guess = [1000, 10000, 1E-6, 0.7]

        circuit = CustomCircuit(circuit, initial_guess=initial_guess)
        print(circuit)

        circuit.fit(frequencies, Z)

        results = circuit.parameters_
        R0 = results[0]
        R1 = results[1]
        Q1 = results[2]
        a1 = results[3]
        print(f'R0 = {R0} ohms')
        print(f'R1 = {R1} ohms')
        print(f'Q1 = {Q1} F.s^(a-1)')
        print(f'a1 = {a1}')

        Z_fit = circuit.predict(frequencies)

        import matplotlib.pyplot as plt
        from impedance.visualization import plot_nyquist

        fig, ax = plt.subplots()
        plot_nyquist(ax, Z, fmt='o')
        plot_nyquist(ax, Z_fit, fmt='-')

        plt.legend(['Data', 'Fit'])
        plt.show()

        Rs[j] = R0
        Rct[j] = R1
        a_hat_1[j] = a1
        C1[j] = Q1/area
        ECSA[j] = (C1[j]*(((1/Rct[j])+(1/Rs[j]))**(a_hat_1[j]-1)))**(1/a_hat_1[j])
        j = j+1

    print(Rct)
    print(time_value)
    print(ECSA)
    fig = plt.figure()
    plt.scatter(time_value,Rct)
    if type == 'GEIS':
        plt.xlabel(f'Time (min) at {current_used} mA/cm2')
        plt.ylabel('Rct (ohm)')
    if type == 'PEIS':
        plt.xlabel('Time (min) at 1.377 V vs. NHE (min)')
        plt.ylabel('Rct (ohm)')
    plt.savefig(f'{root_path}{test_name}_Rct_Time.png')
    plt.show()

    fig = plt.figure()
    plt.scatter(time_value,ECSA)
    if type == 'GEIS':
        plt.xlabel(f'Time (min) at {current_used} mA/cm2')
        plt.ylabel('Capacitance (F/cm2)')
    if type == 'PEIS':
        plt.xlabel('Time (min) at 1.377 V vs. NHE (min)')
        plt.ylabel('Capacitance (F/cm2)')
    plt.savefig(f'{root_path}{test_name}_Cap_Time.png')
    plt.show()

    #Put the data into a .json file format for storage
    dict = {test_name:
                {'GEIS':{'time': time_value.tolist(), 'Rs': Rs.tolist(), 'Rct': Rct.tolist(), 'Cap': C1.tolist(),
                              'a_hat_1': a_hat_1.tolist(), 'ECSA': ECSA.tolist()}}}

    with open(f'{root_path}{test_name}.json', 'w') as fp:
        json.dump(dict, fp, indent = 4)

if __name__ == '__main__':
    fit_EIS_csv(
    root_path='C:\\Users\\Blackr\\Documents\\Data Main\\NRC-M MAPs\\Experimental\\rbnbp153\\TestE\\',
    test_name = 'rbnbp153_TestE',  # Use format 'rbnb2p121_Testx'
    area = .847*.977,  # cm2
    current_used = 0.2,  # mA/cm2
    type = 'GEIS'
    )
```

```python
'''DRT Analysis'''
import numpy as np
import pandas as pd
import os
import sys
import time
import json
import matplotlib.pyplot as plt
%matplotlib inline
import glob
from numpy import genfromtxt
import impedance
from scipy.signal import find_peaks
import scipy
from lmfit import Model
from lmfit.models import GaussianModel, ConstantModel
from lmfit.models import SkewedGaussianModel

from bayes_drt.inversion import Inverter
from bayes_drt import eis_utils as gt

#Output
nb = 'rbnbp153'
test = 'TestA'
test_name = f'{nb}_{test}'
'''Call in data and fit DRT'''
def fit_EIS_csv(root_path,test_name,area,current_used,type):
    import matplotlib.pyplot as plt
    files = [#f'{root_path}{test_name}_Char_01_GEIS_C01.csv',
#              f'{root_path}{test_name}_Char_03_GEIS_C01.csv',
               f'{root_path}{test_name}_Char_05_GEIS_C01.csv',
             f'{root_path}{test_name}_Char_07_GEIS_C01.csv',
             f'{root_path}{test_name}_Char_09_GEIS_C01.csv',
             f'{root_path}{test_name}_Char_11_GEIS_C01.csv',
             f'{root_path}{test_name}_Char_13_GEIS_C01.csv',
#              f'{root_path}{test_name}_Char_15_GEIS_C01.csv',
#              f'{root_path}{test_name}_Char_17_GEIS_C01.csv',
#              f'{root_path}{test_name}_Char_19_GEIS_C01.csv',
             #f'{root_path}{test_name}_Char_21_GEIS_C01.csv',
             ]

    k = 0 # used for the time_value recording 
    time_value = np.zeros(len(files))
    Rs = np.zeros(len(time_value))
    Rct = np.zeros(len(time_value))
    tau_center = np.zeros(len(time_value))
    ECSA = np.zeros(len(time_value))
    
    amplitude = np.zeros(len(time_value))
    center = np.zeros(len(time_value))
    sigma = np.zeros(len(time_value))
    gamma = np.zeros(len(time_value))
    area = np.zeros(len(time_value))
    MSE = np.zeros(len(time_value))
    j = 0 #increment counter only
    
    for i in files:
        print(f'Analyzing data for \n {i}')
        
        data = genfromtxt(i, delimiter=',', skip_header=67)
        data = np.transpose(data)
        frequencies = data[0]
        abs_Ewe = data[11]
        abs_I = data[12]
        Phase_Zwe = data[4]
        Re_Z = data[1]
        Im_Z = -data[2]
        t = np.mean(data[5])/60
        time_value[k] = t
        k = k+1
        
        #For completeness sake, put into a dictionary
        data = {'frequencies':frequencies.tolist(),'abs_Ewe':abs_Ewe.tolist(),'abs_I':abs_I.tolist(),'Phase_Zwe':Phase_Zwe.tolist()}

        abs_Z = np.divide(data['abs_Ewe'], data['abs_I'])
        Re_Z = abs_Z * (np.cos(np.radians(data['Phase_Zwe'])))
        Im_Z = abs_Z * (np.sin(np.radians(data['Phase_Zwe'])))
        #Convert Re_Z and Im_Z into a single complex numpy array
        Z = Re_Z + 1j*Im_Z

        #Below is built in library to cut frequencies below x-axis
#       frequencies, Z = preprocessing.ignoreBelowX(frequencies,Z)
        mask = frequencies < 5000
        frequencies = frequencies[mask]
        Z = Z[mask]
        Re_Z = Re_Z[mask]
        Im_Z = Im_Z[mask]
        #Putting into a dataframe for use with plotting and program
        values = np.array([frequencies,Re_Z,Im_Z])
        labels = ['Freq','Zreal','Zimag']
        Zdf = pd.DataFrame(values.transpose(), columns = labels)

        plt.scatter(Zdf['Zreal'].values,-Zdf['Zimag'].values)
        
        "Fit the data"
        # by default, the Inverter class is configured for the DRT
        inv_hmc = Inverter()
        inv_map = Inverter()

        # Perform HMC fit
        start = time.time()
        inv_hmc.bayes_fit(frequencies,Z)
        elapsed_hmc = time.time() - start
        print('HMC fit time {:.2f}'.format(elapsed_hmc))

        # Perform MAP fit
        start = time.time()
        inv_map.map_fit(frequencies,Z)
        elapsed_map = time.time() - start
        print('MAP fit time {:.2f}'.format(elapsed_map))
        
        "Visualize DRT and impedance fit"
        # plot impedance fit and recovered DRT
        fig,axes = plt.subplots(1,2,figsize=(8,3.5))

        # plot impedance data
        gt.plot_nyquist(Zdf,ax=axes[0],s=20,alpha=0.5,label='Data')

        # plot model fit of impedance
        Z_hmc = inv_hmc.predict_Z(frequencies)
        Z_map = inv_map.predict_Z(frequencies)
        df_hmc = gt.construct_eis_df(frequencies,Z_hmc)
        df_map = gt.construct_eis_df(frequencies,Z_hmc)
        gt.plot_nyquist(df_hmc,ax=axes[0],plot_func='plot',c='k',label='HMC Fit')
        gt.plot_nyquist(df_map,ax=axes[0],plot_func='plot',c='r',label='MAP Fit')

        # evaluate recovered DRT at given tau values
        tau_plot = np.logspace(-5,2,200) # the space overwhich you want to evaluate the DRT 
        g_hmc = inv_hmc.predict_distribution('DRT',eval_tau=tau_plot)
        g_map = inv_map.predict_distribution('DRT',eval_tau=tau_plot)
        g_lo = inv_hmc.predict_distribution('DRT',eval_tau=tau_plot,percentile=2.5)
        g_hi = inv_hmc.predict_distribution('DRT',eval_tau=tau_plot,percentile=97.5)
        # plot recovered DRT
        axes[1].plot(tau_plot,g_hmc,c='k',label='HMC mean')
        axes[1].plot(tau_plot,g_map,c='r',label='MAP')
        axes[1].set_xscale('log')
        # plot the recovered CI
        axes[1].fill_between(tau_plot,g_lo,g_hi,label='95% CI',color='k',alpha=0.15)
        #axes[1].set_ylim(0,3.5)
        axes[1].legend()
        axes[1].set_xlabel(r'$\tau$ / s')
        axes[1].set_ylabel(r'$\gamma \, (\ln{\tau})$ / $\Omega$')

        fig.tight_layout()
        plt.show()

        #Store each raw data file as its own separate .json   
        dict_raw = {test_name:
            {'Raw_DRT':{'tau':tau_plot.tolist() , 'hmc':g_hmc.tolist() , 'map':g_map.tolist(),
                   'hmc_time':elapsed_hmc,'map_time':elapsed_map}}} #Might as well add the impedance data to this for safe keeping
        
        with open(f'{root_path}{test_name}_RawDRT_{j}.json', 'w') as fp: # TODO make name correspond to the .csv file name
            json.dump(dict_raw, fp, indent = 4)
        
        
        ''' post processing goes here'''
        #Step 1 - Find the peaks
        peaks, _ = find_peaks(g_hmc, height=1) #height 1 helps to prevent finding insignificant bumps
        # plt.plot(a_x[peaks], a_y[peaks], "x")
        # plt.plot(a_x,a_y)
        # plt.xscale('log')
        
        #Step 2 - Fit the skewed guassian
        x_array = np.log(tau_plot) # put to log scale
        amp1 = g_hmc[peaks][0] # set initial parameter values
        cen1 = x_array[peaks][0]

        model = SkewedGaussianModel()
        params = model.make_params(amplitude=amp1, center=cen1, sigma=1, gamma=-0)

        result = model.fit(g_hmc, params, x=x_array) # fit the model

#         print(result.fit_report()) 
#         print(f'area = {np.trapz(result.best_fit)}')
        residuals = g_hmc - result.best_fit
        plt.plot(x_array, g_hmc)
        plt.plot(x_array, result.best_fit) 
        plt.plot(x_array, residuals)
        plt.ylim(0,300)
        plt.show()
        
        stuff = [] # blank array to populate
        i = 0 # iterative counter
        for name, param in result.params.items():
#             print('{:7s} {:11.5f} {:11.5f}'.format(name, param.value, param.stderr))
            stuff.append(param.value)
            i+=1
            
        amplitude[j] = stuff[0] #populate the necessary fit arrays
        center[j] = stuff[1]
        sigma[j] = stuff[2]
        gamma[j] = stuff[3]
        area[j] = np.array(np.trapz(result.best_fit)) # calculate the area and record
        MSE[j] = np.sum((g_hmc - result.best_fit)**2)/len(g_hmc)
        print(f'MSE = {MSE[j]}')
        
        #Store fit data into a separate dictionary
        dict_raw = {test_name:
            {'Fit_DRT':{'amplitude':amplitude[j] , 'center':center[j] , 'sigmap':sigma[j],
                   'gamma':gamma[j],'area':area[j],'MSE':MSE[j]}}} #Might as well add the impedance data to this for safe keeping
        
        with open(f'{root_path}{test_name}_FitDRT_{j}.json', 'w') as fp: # TODO make name correspond to the .csv file name
            json.dump(dict_raw, fp, indent = 4)

        ''' TODO - Next would be calculating the metric information, once determine how to do '''
        Rs[j] = 1
        Rct[j] = 1
        ECSA[j] = 1
        tau_center[j] = 1
        #put into metric_DRT.json for individual file
        #MAKE SURE THIS STAYS AS THE LAST LINE
        j = j+1 #counter increase

    '''Here goes plots of the post-processed data'''
        
    plt.plot(time_value,center)
    plt.show()
    plt.plot(time_value,area)
    plt.show()
    
    dict = {test_name:
                {'DRT':{'time': time_value.tolist(), 'Rs': Rs.tolist(), 'Rct': Rct.tolist(),
                        'tau_center':tau_center.tolist(),'ECSA': ECSA.tolist()}}}

    with open(f'{root_path}{test_name}_DRT.json', 'w') as fp:
        json.dump(dict, fp, indent = 4)

if __name__ == '__main__':
    fit_EIS_csv(
    root_path=f'C:\\Users\\Blackr\\Documents\\Data Main\\NRC-M MAPs\\Experimental\\{nb}\\{test}\\',
    test_name = test_name,  # Use format 'rbnb2p121_Testx'
    area = .92*.985,  # cm2
    current_used = 0.2,  # mA/cm2
    type = 'GEIS'
    )

# dict = {'test':
#                 {'DRT':
#                      {'Raw_DRT':{'n_step':{'xxx':'hello'}},
#                      'Fit_DRT':'',
#                      'metric':''}}}
```
